# EDA

I have 20 time series datasets. I will look into the plots. Discover trends, seasonality, and residuals. I will plot ACF and PACF to figure out AR and MA. My final model is SARIMA model.

### Import Packages

```python
import matplotlib.pyplot as plt
import pickle
from statsmodels.graphics.tsaplots import plot_acf ,plot_pacf
import pandas as pd
```

### Load Time Series Datasets

```python
with open('../data/final/dataset_dict.pickle', 'rb') as data:
    dataset_dict = pickle.load(data)
```

### Time Series Plots - Top 10 counts

```python
fig, ax = plt.subplots(nrows=5, ncols=2, figsize=(15, 15), constrained_layout=True)
for genre, i in zip(dataset_dict.keys(), range(10)):
    r = i%5
    c = i//5
    dataset_dict[genre].plot(ax=ax[r][c], title=f'{i+1}', color='g');
```

### Time Series Plots - Top 10 growth of top 10% counts

```python
growth_dict = {}
for (genre, df), i in zip(dataset_dict.items(), range(50)):
    growth = (df.tail(1).values-df.mean().values)/df.mean().values
    growth_dict[genre] = growth
```

```python
top10_growth = sorted(growth_dict.items(), key=lambda x: x[1], reverse=True)[:10]
```

```python
fig, ax = plt.subplots(nrows=5, ncols=2, figsize=(15, 15), constrained_layout=True)
for genre, i in zip(dict(top10_growth).keys(), range(10)):
    r = i%5
    c = i//5
    dataset_dict[genre].plot(ax=ax[r][c], title=f'{i+1}', color='r');
```

### Save Data

```python
top10_counts_keys = list(dict(list(dataset_dict.items())[:10]).keys())
top10_growth_keys = list(dict(top10_growth).keys())
top10_counts_dict = {key: dataset_dict[key] for key in top10_counts_keys}
top10_growth_dict = {key: dataset_dict[key] for key in top10_growth_keys}
```

```python
# Save top10_counts_dict
with open('../data/final/top10_counts_dict.pickle', 'wb') as dict_:
    pickle.dump(top10_counts_dict, dict_)
    
# Save top10_growth_dict
with open('../data/final/top10_growth_dict.pickle', 'wb') as dict_:
    pickle.dump(top10_growth_dict, dict_)
```
