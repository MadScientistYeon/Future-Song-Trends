# Modeling
I train models for top 100 artists. Future growth is caculated with each models and saved as dictionary.

### Import Packages

```python
from train_test import train_test
from GRU_model import GRU_model
from actual_pred_plot import actual_pred_plot
import pickle
```

### Load Data

```python
with open('../data/final/artists_dict.pickle', 'rb') as artists:
    artists_dict = pickle.load(artists)
    artists.close()

with open('../data/final/rank_df.pickle', 'rb') as rank:
    rank_df = pickle.load(rank)
    rank.close()
```

### Modeling

```python
future_growth = {}
history_dict = {}
```

```python
for artist in rank_df.index:
    df = artists_dict[artist].resample('W').mean().dropna()
    X_train, y_train, X_test, sc = train_test(df, 5, 13)
    my_GRU_model, GRU_prediction, history = GRU_model(X_train, y_train, X_test, sc, artist)
    my_GRU_model.save('../models/{}.h5'.format(artist))
    future_growth[artist] = (GRU_prediction[-1][12] - GRU_prediction[-1][0])/GRU_prediction[-1][0]
    history_dict[artist] = history.history
```

```python
with open('../data/final/future_growth.pickle', 'wb') as growth:
    pickle.dump(future_growth, growth)
    growth.close()
```

```python
with open('../data/final/history.pickle', 'wb') as history:
    pickle.dump(history_dict, history)
    growth.close()
```
